```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_joel = pd.read_csv('datos_optimizacion/datos_joel.txt', header=0, delimiter=",", names=['h', 'n_int', 'time', 'run_time'])
data_windows = pd.read_csv('datos_optimizacion/datos_windows.txt', header=0, delimiter=",", names=['h', 'n_int', 'time', 'run_time'])
data_mac = pd.read_csv('datos_optimizacion/datos_mac.txt', header=0, delimiter=",", names=['h', 'n_int', 'time', 'run_time'])

fig=plt.figure(figsize=(13,7)) #Size of the plot
ax=fig.add_subplot(111)

plt.plot(np.log(data_joel.n_int), np.log(data_joel.run_time), label="Joel", color='tab:orange')
plt.plot(np.log(data_windows.n_int), np.log(data_windows.run_time), label="Windows", color='tab:blue')
plt.plot(np.log(data_mac.n_int), np.log(data_mac.run_time), label="Mac", color='tab:purple')
plt.xlabel("Nmero de interaciones", fontsize=14)
plt.ylabel("Tiempo de ejecucin", fontsize=14)
plt.legend(fontsize=14)
plt.title("Relacin entre el nmero de interaciones y el tiempo de ejecucin para las \n distintas mquina, en escala logartmica", fontsize=16)
plt.grid()

plt.savefig('latex/plots/optimizacion.png', bbox_inches='tight', pad_inches=0.1)
```
